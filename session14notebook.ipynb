## Goal 4
Find the largest group of car makes for each gender

```python
male_list = filter(lambda c: c.gender == 'Male', mylist)
female_list = filter(lambda c: c.gender == 'Female', mylist)
```

```python
male_vehicle_make_dict = defaultdict(int)
    
for row in list(male_list):
    male_vehicle_make_dict[row.vehicle_model] += 1

all_values = male_vehicle_make_dict.values()
max_value = max(all_values)
malenewDict = filterTheDict(male_vehicle_make_dict, lambda elem: elem[1] == max_value)
```

```python
female_vehicle_make_dict = defaultdict(int)
    
for row in list(female_list):
    female_vehicle_make_dict[row.vehicle_model] += 1

all_values = female_vehicle_make_dict.values()
max_value = max(all_values)
femalenewDict = filterTheDict(female_vehicle_make_dict, lambda elem: elem[1] == max_value)
```

```python
print("max vehicle make by male " + str(malenewDict))
print("max vehicle make by Female " + str(femalenewDict))
```
